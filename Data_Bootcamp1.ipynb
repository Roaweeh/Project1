```python
# Extract the appointment hour from the 'AppointmentDay' column and add 'Hour' column
df['Hour']=df.ScheduledDay.str.slice(-9,-7)
```

```python
#Change the 'ScheduledDay' and 'AppointmentDay' to datetime. change the 'Hour' to integer
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.date.astype('datetime64[ns]')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date.astype('datetime64[ns]')
df['Hour']=df['Hour'].astype(int)
```

```python
# Relation between gender and attendances to the hospital appointment
sns.countplot(x=df['Absent'], hue=df['Gender']);
plt.title('Male vs Female Attendance');
```

```python
# Attendance in terms of the Neighborhood
plt.figure(figsize=(20,10))
sns.countplot(x=df.Neighbourhood);
plt.title('Attendance by Neighborhood')
plt.xticks(rotation=90);
```

```python
# No show hospital appointment in relation to the neighborhood
plt.figure(figsize=(10, 12))
sns.countplot(y=df['Neighbourhood'], hue=df['Absent'], hue_order=['Yes'], order=df.Neighbourhood.value_counts().index[::-1]);
plt.xticks(rotation=45);
plt.title('Absenteeism in Different Neighborhood');
```

```python
df['Age'].hist(bins=10);
```

```python
# this trick makes age easier as I divided them into segments to make #it easier visualizing
df['Age'] =[round(a,-1) for a in df['Age']]
df['Age'].value_counts()
```

```python
# Relation between Absenteeism and Age
plt.figure(figsize=(20,5))
a=sns.countplot(x=df['Age'], hue=df['Absent'], hue_order=['Yes'])
for label in a.containers:
  a.bar_label(label)
plt.title('Absenteeism by Age');
```

### Hospital appointments are more likely to be missed by those between the ages of 20 and 40

```python
# Relation between Absenteeism and SMS sent
plt.figure(figsize=(5,5))
b=sns.countplot(x=df['Absent'], hue=df['SMS_received'])
for label in b.containers:
  b.bar_label(label)
plt.title('Absenteeism Vs SMS');
```

### Attendance at the hospital appointment does not appear to be affected by whether the SMS was received or not

```python
df.groupby(['Absent'])[['Hypertension',
       'Diabetes', 'Alcoholism', 'Handicap']].sum()
```

```python
# Relation between Absenteeism in different diseases
df.groupby(['Absent'])[['Hypertension',
       'Diabetes', 'Alcoholism', 'Handicap']].sum().plot(kind='bar')
plt.xticks(rotation=45);
plt.title('Absenteeism in Different Diseases')
```

### Hypertensive patients miss their hospital appointments more frequently than diabetics, alcoholics, and handicapped patients

```python
# Relation between Absenteeism and the Hour in the day
plt.figure(figsize=(15,7))
b=sns.countplot(x=df['Hour'], hue=df['Absent'],hue_order=['Yes'])
for label in b.containers:
  b.bar_label(label)
plt.title('Absenteeism by Hour')
```

### Appointments between 7 and 10 am are less likely to be attended

```python
# Relation between Absenteeism and the Day in the Week
DayOfWeek=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
plt.figure(figsize=(10,5))
sns.countplot(x=df['Day'], hue=df['Absent'], hue_order=['Yes'], order=DayOfWeek)
plt.title('Absenteeism by Day')
```

### The most commonly missed days for patients are Tuesday and Wednesday, where patients are more likely to attend their hospital appointment on Thursday than any other day of the week.


```python
# Relation between Absenteeism and the Month in the year
plt.figure(figsize=(10,5))
b=sns.countplot(x=df['Month'], hue=df['Absent'],hue_order=['Yes'])
for label in b.containers:
  b.bar_label(label)
plt.title('Absenteeism by Month')
```
